### CNN 6 Training
Dataset:
* 13 folders (A~L and X)
    * A-L: 5400 tiff images
        * 900 malware samples per family
    * X: 7200 tiff images (unlabeled samples)
        * 1200 unlabeled malware samples
* ALL IMAGES FROM MALWARE SAMPLES (6 total distinct image type)
    * each image: 224 x 224 (size)
* Format (6 images for a given sample denoted a thru f): 
    * Family A, sample 000: A_000_a.tiff, A_000_b.tiff, ..., A_000_f.tiff 
    * Family A, sample 001: A_001_a.tiff, A_001_b.tiff, ..., A_001_f.tiff 
    * Family B, sample 000: B_000_a.tiff, B_000_b.tiff, ..., B_000_f.tiff 
    * etc

#### Step 1
Use labeled samples in folders A thru L
1. Train 6 CNN models
    * One model for each image type a thru f
2. Tune hyperparameters independently for each of the 6 models
3. 80:20 train:test split for all 
    * Don't need cross validation

Goal: Maximize test accuracy for each model
* Clearly state 
    * hyperparameters tested
    * hyperparameters selected
    * best accuracy obtained
    * accuracy and loss graph
